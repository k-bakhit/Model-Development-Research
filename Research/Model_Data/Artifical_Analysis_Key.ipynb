```python
import os, requests, json
from dotenv import load_dotenv
```

```python

load_dotenv("Artificial_Analysis_Key.env")
key = os.environ["Artificial_Analysis_Key"]
print("Key loaded:", bool(key))
```

```python
r = requests.get(
    "https://artificialanalysis.ai/api/v2/data/llms/models",
    headers={"x-api-key": key},
)
r.raise_for_status()
payload = r.json()
print("Status:", r.status_code, "| models:", len(payload.get("data", [])))
```

```python
with open("artificial_analysis.json", "w") as f:
    json.dump(payload, f, indent=2)
print("Saved artificial_analysis.json")
```
